# Types of Missing Data:

# 2) Missing not at random (MNAR) 

## or Systematic Missing Values

(MNAR) (also known as nonignorable nonresponse) is 
### data that is neither MAR nor MCAR 

### (i.e. the value of the variable that's missing is related to the reason it's missing)
To extend the previous example, this would occur if men failed to fill in a depression survey because of their level of depression

```python
import pandas as pd
#Read Titanic Training Data Set
df = pd.read_csv('C:/Users/Kapil Wanaskar/Data Sets/titanic_train (1).csv')
df
```

```python
df['cabin'].isnull()
```

```python
import numpy as np
df['cabin_null'] = np.where(df['cabin'].isnull(),1,0) #Null, Nan values are replaced by 1 in new column "cabin_null"
df['cabin_null']
```

```python
df.columns
# new column 'cabin_null' is generated
```

```python
# find the percentage of null values  = 659/850 %
df['cabin_null'].mean()
#mean = sum / total
```

```python
#comparing "Cabin_null" vs survived
df.groupby(['survived'])['cabin_null'].mean()
# 0 ~ Not Survived --> 88% have missing values
# 1 ~ Survived     --> 59% have missing values
#its obvious that,
# missing values for "not suvived" people should be greater than missing values for "Suvived" people
```

# 3) Missing at random (MAR) 

Missing at random (MAR) occurs when the missingness is not random, but where missingness can be fully accounted for by variables where there is complete information. Since MAR is an assumption that is impossible to verify statistically, we must rely on its substantive reasonableness.
